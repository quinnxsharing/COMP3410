The rationale for the last excercise is to indicate how to model the path-planning in a grid with obstacles. It is more efficient to create **Grid** class. Complete the following. 

```python
from collections import defaultdict

class Grid:
    ##class constructor with a parameter a (numpy) array, 2 integer tuples for start and goal respectively
    def __init__(self, binary_grid_array, start, goal):
        self.grid = binary_grid_array #the grid of the grid-world
        self.size = binary_grid_array.shape # A numpy 2D array is given as input
        self.start = start
        self.goal = goal
        self.neighbour = defaultdict(list) #the dictionary containing the list of neighbours, dynamically generated
    
    # The following position is crucial as it defines the actions and hence the graph! 
    # The parameter current_pos is a tuple giving the current coordinates on the grid. 
    # The list of neighbors will depend upon the grid.
    def neighbours(self, current_pos):
        current_x = current_pos[0]
        current_y = current_pos[1]
        
        # Updates the neighbor dictionary by updating the list of neighbors of curr_pos.
        # Replace "pass" with your code.
        
        right_node = (current_x + 1, current_y)
        if right_node[0] < self.size[0] and self.grid[right_node] == 1:
            self.neighbour[current_pos].append(right_node) if right_node not in self.neighbour[current_pos] else None
        
        left_node = (current_x - 1, current_y)
        if left_node[0] >= 0 and self.grid[left_node] == 1:
            self.neighbour[current_pos].append(left_node) if left_node not in self.neighbour[current_pos] else None
    
        down_node = (current_x, current_y + 1)
        if down_node[1] < self.size[1] and self.grid[down_node] == 1:
            self.neighbour[current_pos].append(down_node) if down_node not in self.neighbour[current_pos] else None
        
        up_node = (current_x, current_y - 1)
        if up_node[1] >= 0 and self.grid[up_node] == 1:
            self.neighbour[current_pos].append(up_node) if up_node not in self.neighbour[current_pos] else None
        
        diag_node1 = (current_x + 1, current_y + 1)
        if diag_node1[0] < self.size[0] and diag_node1[1] < self.size[1] and self.grid[diag_node1] == 1:
            self.neighbour[current_pos].append(diag_node1) if diag_node1 not in self.neighbour[current_pos] else None
        
        diag_node2 = (current_x - 1, current_y + 1)
        if diag_node2[0] >= 0 and diag_node2[1] < self.size[1] and self.grid[diag_node2] == 1:
            self.neighbour[current_pos].append(diag_node2) if diag_node2 not in self.neighbour[current_pos] else None
        
        diag_node3 = (current_x + 1, current_y - 1)
        if diag_node3[0] < self.size[0] and diag_node3[1] >= 0 and self.grid[diag_node3] == 1:
            self.neighbour[current_pos].append(diag_node3) if diag_node3 not in self.neighbour[current_pos] else None
        
        diag_node4 = (current_x - 1, current_y - 1)
        if diag_node4[0] >= 0 and diag_node4[1] >= 0 and self.grid[diag_node4] == 1:
            self.neighbour[current_pos].append(diag_node4) if diag_node4 not in self.neighbour[current_pos] else None

# Create an instance of the Grid class
#grid = Grid(binary_grid_array, start, goal)
```
